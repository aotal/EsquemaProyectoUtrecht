```python
def make_markdown_table(array):

    """ Input: Python list with rows of table as lists
               First element as header. 
        Output: String to put into a .md file 
        
    Ex Input: 
        [["Name", "Age", "Height"],
         ["Jake", 20, 5'10],
         ["Mary", 21, 5'7]] 
    """
    markdown="<ul>"
    for e in array:
        markdown += '<li>' + str(e)+'</li>'
        #markdown +=" <br> "
    markdown+='</ul>'
    return markdown
```
